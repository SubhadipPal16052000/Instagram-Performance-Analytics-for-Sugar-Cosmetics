```python
import os
from pathlib import Path

INPUT_CSV = Path(r"E:\LABMENTICX_PROJECT\PROJECT_10\sugar_cosmetics_instagram_data.csv")
CLEANED_CSV = Path(r"E:\LABMENTICX_PROJECT\PROJECT_10\cleaned_sugar_instagram.csv")
HASHTAGS_CSV = Path(r"E:\LABMENTICX_PROJECT\PROJECT_10\cleaned_sugar_instagram_hashtags.csv")
```

```python
# Load safely
df_raw = pd.read_csv(INPUT_CSV, dtype=str, keep_default_na=False, na_values=['', 'NA', 'NaN', 'None'])
# Normalize column names
df_raw.columns = [c.strip().lower().replace(' ', '_').replace('-', '_') for c in df_raw.columns]
```

```python
# Map expected columns to canonical names if present
col_map = {}
candidates = {
    'post_date': ['post_date','date','published_at','created_at'],
    'post_id': ['post_id','id','postid'],
    'post_type': ['post_type','type','content_type'],
    'likes': ['likes','like_count','likes_count'],
    'comments': ['comments','comment_count','comments_count'],
    'shares': ['shares','share_count'],
    'saves': ['saves','save_count','bookmarks'],
    'reach': ['reach','audience_reach'],
    'impressions': ['impressions','impression_count'],
    'hashtags': ['hashtags','tags','caption_hashtags'],
    'caption': ['caption','post_text','description','text']
}
for std, cand_list in candidates.items():
    for cand in cand_list:
        if cand in df_raw.columns:
            col_map[cand] = std
            break
```

```python
# Build final df and ensure canonical columns exist
df = df_raw.rename(columns={v:k for k,v in col_map.items()})
for std in candidates.keys():
    if std not in df.columns:
        df[std] = pd.NA
```

```python
# Ensure post_id exists
if 'post_id' not in df.columns or df['post_id'].isna().all():
    df['post_id'] = ["post_" + str(i) for i in range(1, len(df)+1)]
```

```python
# Parse dates
df['post_date'] = pd.to_datetime(df['post_date'], errors='coerce', infer_datetime_format=True)
df['month_year'] = df['post_date'].dt.to_period('M').astype(str)
df['weekday'] = df['post_date'].dt.day_name()
df['hour'] = df['post_date'].dt.hour
```

```python
# Coerce numeric columns
num_cols = ['likes','comments','shares','saves','reach','impressions']
for c in num_cols:
    if c not in df.columns:
        df[c] = pd.NA
    df[c] = df[c].astype(str).str.replace(r'[^\d\.-]', '', regex=True)
    df[c] = pd.to_numeric(df[c], errors='coerce').astype('Int64')
```

```python
# Fill engagement parts with 0 when partially present
for c in ['likes','comments','shares','saves']:
    if not df[c].isna().all():
        df[c] = df[c].fillna(0).astype('Int64')
```

```python
# Normalize post_type
df['post_type_clean'] = df['post_type'].astype(str).str.strip().str.lower().fillna('unknown')
def map_post_type(x):
    x = str(x).lower()
    if 'reel' in x or 'video' in x:
        return 'video'
    if 'carous' in x or 'album' in x:
        return 'carousel'
    if 'story' in x:
        return 'story'
    if 'image' in x or 'photo' in x:
        return 'image'
    return 'other'
df['post_type_clean'] = df['post_type_clean'].apply(map_post_type)
```

```python
# Engagements and engagement rate (prefer reach, fallback to impressions)
df['engagements'] = df[['likes','comments','shares','saves']].sum(axis=1).astype('Int64')
def compute_er(row):
    num = row['engagements']
    if pd.isna(num):
        return pd.NA
    if pd.notna(row['reach']) and row['reach']>0:
        return (int(num)/int(row['reach']))*100
    if pd.notna(row['impressions']) and row['impressions']>0:
        return (int(num)/int(row['impressions']))*100
    return pd.NA
df['engagement_rate_pct'] = df.apply(compute_er, axis=1)
```

```python
# Extract hashtags (simple regex) and explode
import re
def extract_hashtags_text(s):
    if pd.isna(s):
        return []
    s = str(s)
    return [h.lower() for h in re.findall(r'#([A-Za-z0-9_]+)', s)]
```

```python
# prefer explicit hashtags column, else caption
df['hashtags_source'] = df.apply(lambda r: r['hashtags'] if pd.notna(r.get('hashtags')) and str(r.get('hashtags')).strip() else r.get('caption'), axis=1)
df['hashtags_list'] = df['hashtags_source'].apply(extract_hashtags_text)
df['num_hashtags'] = df['hashtags_list'].apply(len).astype('Int64')

df_hashtags = df[['post_id','post_date','post_type_clean','engagements','engagement_rate_pct','hashtags_list']].explode('hashtags_list').rename(columns={'hashtags_list':'hashtag'})
df_hashtags = df_hashtags[df_hashtags['hashtag'].notna() & (df_hashtags['hashtag']!='')].copy()
```

```python
# Deduplicate by post_id keeping most complete
if df['post_id'].duplicated().any():
    df['completeness'] = df[num_cols + ['caption','post_date']].notna().sum(axis=1)
    df = df.sort_values('completeness', ascending=False).drop_duplicates('post_id', keep='first').drop(columns=['completeness'])
```

```python
# Export cleaned files
df.to_csv(CLEANED_CSV, index=False)
df_hashtags.to_csv(HASHTAGS_CSV, index=False)
print("Cleaned files:", CLEANED_CSV, HASHTAGS_CSV)
```

```python
# ---------- Final quick validations & summary
print("Rows loaded:", len(df_raw), "-> cleaned rows:", len(df))
print("Unique posts:", df['post_id'].nunique())
print("Missing post_date:", df['post_date'].isna().sum())
print("Numeric columns non-null counts:")
for c in num_cols:
    print(f"  {c}: {df[c].notna().sum()}")

print("Top post_type_clean counts:")
print(df['post_type_clean'].value_counts(dropna=False).head(10))

print("Top hashtags count (sample):")
if not df_hashtags.empty:
    print(df_hashtags['hashtag'].value_counts().head(10))
else:
    print("  No hashtags extracted.")

# ---------- Export cleaned files
df.to_csv(CLEANED_CSV, index=False)
df_hashtags.to_csv(HASHTAGS_CSV, index=False)
print("Wrote:", CLEANED_CSV, HASHTAGS_CSV)
```

#### VISULIZATIONS


```python
# 1) Monthly Posting Trend
df['month_year'] = pd.to_datetime(df['post_date']).dt.to_period('M').astype(str)
monthly = df.groupby('month_year').size().reset_index(name='post_count').sort_values('month_year')

plt.figure(figsize=(10,4));
plt.plot(monthly['month_year'], monthly['post_count'], marker='o');
plt.xticks(rotation=45, ha='right')
plt.title('Monthly Posting Trend');
plt.xlabel('Month-Year');
plt.ylabel('Number of Posts');
plt.grid(True);
plt.tight_layout()
plt.show()
display(monthly)
```

```python
# 2) Top 10 Posts by Likes
top10_likes = df[['post_id','post_date','post_type_clean','likes','caption']].sort_values('likes', ascending=False).head(10)[::-1]
plt.figure(figsize=(10,6));
plt.barh(top10_likes['post_id'].astype(str), top10_likes['likes']);
plt.title('Top 10 Posts by Likes'); 
plt.xlabel('Likes'); plt.tight_layout()
plt.show()
display(top10_likes)
```

```python
# 3) Top 10 by Engagement Rate (filter reach>100)
filtered = df[df['reach'].fillna(0) > 100].copy()
top10_er = filtered[['post_id','post_date','post_type_clean','engagement_rate_pct','reach','engagements']].sort_values('engagement_rate_pct', ascending=False).head(10)[::-1]
plt.figure(figsize=(10,6));
plt.barh(top10_er['post_id'].astype(str), top10_er['engagement_rate_pct']);
plt.title('Top 10 Posts by Engagement Rate (Reach>100)'); 
plt.xlabel('Engagement Rate (%)');
plt.tight_layout()
plt.show()
display(top10_er)
```

```python
# 4) Average Engagement Rate by Post Type
er_by_type = df.groupby('post_type_clean')['engagement_rate_pct'].mean().reset_index().sort_values('engagement_rate_pct', ascending=False)
plt.figure(figsize=(8,5));
plt.bar(er_by_type['post_type_clean'], er_by_type['engagement_rate_pct']);
plt.title('Avg Engagement Rate by Post Type');
plt.xlabel('Post Type');
plt.ylabel('Engagement Rate (%)');
plt.tight_layout()
plt.show()
display(er_by_type)
```

```python
# 5) Average Engagements by Post Type
eng_by_type = df.groupby('post_type_clean')['engagements'].mean().reset_index().sort_values('engagements', ascending=False)

plt.figure(figsize=(8,5)); 
plt.bar(eng_by_type['post_type_clean'], eng_by_type['engagements']);
plt.title('Avg Engagements by Post Type');
plt.xlabel('Post Type');
plt.ylabel('Avg Engagements');
plt.tight_layout()
plt.show()
display(eng_by_type)
```

```python
# 6) Avg Engagement Rate by Weekday (Mon-Sun)
df['weekday'] = pd.to_datetime(df['post_date']).dt.day_name()
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
er_by_weekday = df.groupby('weekday')['engagement_rate_pct'].mean().reindex(order).reset_index()

plt.figure(figsize=(9,5)); 
plt.bar(er_by_weekday['weekday'], er_by_weekday['engagement_rate_pct']); 
plt.title('Avg Engagement Rate by Weekday'); 
plt.xlabel('Weekday'); 
plt.ylabel('Engagement Rate (%)');
plt.tight_layout()
plt.show()
display(er_by_weekday)
```

```python
# 7) Engagement Funnel (aggregated totals)
sum_impr = df['impressions'].sum(min_count=1)
sum_reach = df['reach'].sum(min_count=1)
sum_likes = df['likes'].sum(min_count=1);
sum_comments = df['comments'].sum(min_count=1)
sum_shares = df['shares'].sum(min_count=1);
sum_saves = df['saves'].sum(min_count=1)
labels = [];
vals = []
if not np.isnan(sum_impr): labels.append('Impressions'); vals.append(sum_impr)
if not np.isnan(sum_reach): labels.append('Reach'); vals.append(sum_reach)
labels.extend(['Likes','Comments','Shares','Saves']); vals.extend([sum_likes,sum_comments,sum_shares,sum_saves])


plt.figure(figsize=(8,5)); 
plt.bar(labels, vals); 
plt.title('Engagement Funnel (Aggregated Totals)'); 
plt.tight_layout()
plt.show()
for l,v in zip(labels,vals): print(l, v)
```

```python
# 9) Likes vs Reach scatter with linear fit
scatter = df[df['likes'].notna() & df['reach'].notna()].copy()
scatter = scatter[(scatter['reach']>0) & (scatter['likes']>=0)]
x = scatter['reach'].values; y = scatter['likes'].values

plt.figure(figsize=(8,6));
plt.scatter(x,y, alpha=0.5, s=10);
plt.xlabel('Reach');
plt.ylabel('Likes');
plt.title('Likes vs Reach (scatter)')
if len(x)>1:
    coeffs = np.polyfit(x,y,1);
    p = np.poly1d(coeffs);
    xs = np.linspace(np.min(x), np.max(x), 100);
    plt.plot(xs, p(xs), linewidth=2)
    # R^2
    y_pred = p(x);
    ss_res = np.sum((y - y_pred)**2);
    ss_tot = np.sum((y - np.mean(y))**2);
    r2 = 1 - ss_res/ss_tot if ss_tot>0 else np.nan
    plt.text(0.05,0.95, f'y={coeffs[0]:.6f}x+{coeffs[1]:.1f}\nR^2={r2:.4f}', transform=plt.gca().transAxes, verticalalignment='top')
    print('Linear fit coeffs:', coeffs, 'R2:', r2)
plt.tight_layout();
plt.show()
display(scatter)
```
